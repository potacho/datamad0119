#### 4. Import the users table

#### 5. Rename Id column to userId

```python
users = users.rename(columns={'Id':'userId'})
users.head()
```

#### 6. Import the posts table. 

```python
posts = pd.read_sql_query('SELECT * FROM stats.posts', engine)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts = posts.rename(columns={'Id':'postId','OwnerUserId':'userId'})
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users2 = users[['userId','Reputation','Views','UpVotes','DownVotes']].copy()
posts2 = posts[['postId','Score','userId','ViewCount','CommentCount']].copy()
display(users2.head())
display(posts2.head())
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
users_post = pd.merge(users2, posts2, on='userId')
users_post.head()
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = users_post.isnull().sum()
display(null_cols[null_cols > 0])
#Existen 48396 valores nulos, todos ellos en la columna 'ViewCount'
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
drop_cols = list(null_cols[null_cols > 0].index)
display(drop_cols)
users_post_nn = users_post.drop(drop_cols, axis=1)
users_post_nn.head()
#Al ser solo una columna la que tiene los valores nulos y ser todos los valores de la columna nulos,
#la opción más razonable es eliminar toda la columna.
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
display(users_post_nn.dtypes)
users_post_nn['userId'] = users_post_nn['userId'].astype('object')
display(users_post_nn.dtypes)
#Parece oportuno cambiar el tipo de dato de la columna 'userId' ya que puede se trata de un identificador 
#que en algún momento pudiese ser definido como un str, además de que no nos diría nada de las estadisticas.
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
analysis = users_post_nn.copy()
stats = analysis.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
display(stats)

outliers = pd.DataFrame(columns=users_post_nn.columns)
for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 3
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = users_post_nn[(users_post_nn[col] < lower) | 
                   (users_post_nn[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)

display(outliers)

outliers.to_csv('/home/potacho/github/datamad0119/module-1/lab-data_cleaning/your-code/outliers.csv', index=False)
#Pendiente de terminar y analizar...
```

# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
w = np.exp(x)
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot('211')
plt.plot(x,y,'m')
plt.title('Función lineal')
plt.subplot('212')
plt.plot(x,z,'g')
plt.title('Función cuadrática')
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
plt.subplot('211')
plt.plot(x,y,'m')
plt.title('Función lineal vs Función cuadrática')
plt.plot(x,z,'g')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
plt.figure(figsize=(10, 10), dpi= 80, facecolor='w')
plt.subplot('211')
plt.plot(x,y,'m')
plt.title('Función lineal')
plt.subplot('212')
plt.plot(x,z,'g')
plt.title('Función cuadrática')


plt.figure(figsize=(10, 100), dpi= 80, facecolor='w')
plt.subplot('211')
plt.plot(x,y,'m')
plt.title('Función lineal vs Función cuadrática')
plt.plot(x,z,'g')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
plt.figure(figsize=(10, 10), dpi= 80, facecolor='w')
plt.subplot('211')
plt.plot(x,y,'m')
plt.title('Función lineal')
plt.ylabel(r'$y=2x$')
plt.subplot('212')
plt.plot(x,z,'g')
plt.title('Función cuadrática')
plt.ylabel(r'$z=x^2$')
plt.xlabel('Valores de x')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

plt.figure(figsize=(10, 10), dpi= 80, facecolor='w')
plt.xlabel('Valores de x')
plt.ylabel('Valores de w/z')
plt.yscale('log')
plt.plot(x,w,'m')
plt.plot(x,z,'g')
plt.legend([r'$w=e^x$',r'$z=x^2$'])
```
